```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import time
import warnings
from skinet.utils.athena_tools import *
warnings.filterwarnings("ignore")
```

```python
conn = athena_connection()
df = load_from_athena(conn, '''
WITH t1 AS (
SELECT ctnum, ins_type_nm,  MAX(dt) AS dt
  FROM sap_data.zsdr6051
GROUP BY ctnum, ins_type_nm
)

SELECT t1.ctnum, t2.ins_stat_nm, t2.ins_type_nm, t2.ins_amt, t2.ins_amt_type_nm, t2.ins_user4, t2.valid_sdate, t2.valid_edate, t1.dt
  FROM t1
LEFT JOIN (SELECT * FROM sap_data.zsdr6051) t2
       ON t1.ctnum = t2.ctnum
      AND t1.ins_type_nm = t2.ins_type_nm 
      AND t1.dt = t2.dt
''')
```
