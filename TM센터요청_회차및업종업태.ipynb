```python
from skinet.utils.athena_tools import *
conn = athena_connection()
tmresult = load_from_athena(conn, '''
 SELECT DISTINCT a.ctnum AS ctnum_hash
      , a.kunnr AS kunnr_hash
	  , d.ctnum
	  , d.kunnr
	  , a.first_bildate
	  , b.j_1kftbus
	  , b.j_1kftind
	  , c.dialresultcd
	  , c.nth_contact_ctnum
	  , c.nth_contact_kunnr
	  , c.nth_contact_first_date_ctnum
	  , c.nth_contact_final_date_ctnum
	  , c.prior_contact_final_date_ctnum
	  , c.prior_contact_final_date_kunnr
  FROM sap.zsdt4013 a
LEFT JOIN (SELECT kunnr, j_1kftind, j_1kftbus FROM sap.kna1) b
      ON a.kunnr = b.kunnr
LEFT JOIN (SELECT * FROM default.tm_period_final) c
      ON a.ctnum = c.s25csq
	 AND a.kunnr = c.kunnr
LEFT JOIN (SELECT kunnr, ctnum, kunnr_hash, ctnum_hash FROM adt_work.hash_info) d
      ON a.ctnum = d.ctnum_hash
	 AND a.kunnr = d.kunnr_hash
WHERE itype_l = '1'
''') 
```
